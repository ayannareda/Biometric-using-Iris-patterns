```python
import os
import numpy as np
```

### Importing Dataset

```python
# Infrared images from Utiris V.1 dataset is used
# If doesnt exist download from (https://utiris.wordpress.com/).
```

```python
# Load the data in a dictionary with two keys
# data -> contains image paths
# targets -> Image label
```

```python
def load_data(path):
    data = []
    target = []
    target_i = 0
    index_used = False
    for dirpath, dirnames, filenames in os.walk(path):
        for f in filenames:
            if f.endswith('.bmp'):
                data.append('{}\{}'.format(dirpath, f))
                target.append(target_i)
                index_used = True
        if index_used:
            target_i += 1
            index_used = False
    print ("Dataset Loaded")
    return {'data': np.array(data), 'target': np.array(target)}
```
